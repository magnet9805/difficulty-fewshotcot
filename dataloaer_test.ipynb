```python
import pandas as pd
import ast

# txt 파일을 읽어옵니다.
un = pd.read_table('./results/wrong_gsm8k.txt', header=None)
# 각 행의 문자열을 딕셔너리로 변환하고 새로운 데이터프레임 생성
un = pd.DataFrame(un[0].apply(ast.literal_eval).tolist())

# txt 파일을 읽어옵니다.
di = pd.read_table('./results/wrong_gsm8k_difficulty_ver1.txt', header=None)
# 각 행의 문자열을 딕셔너리로 변환하고 새로운 데이터프레임 생성
di = pd.DataFrame(di[0].apply(ast.literal_eval).tolist())

unli = set(un.idx)
dili = set(di.idx)
print(len(unli-dili))
print(len(dili-unli))
print(len(unli & dili))
```
